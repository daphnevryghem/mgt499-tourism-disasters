## Visualization

- Display a heatmap of the correlation matrix between the columns: 'hg/ha_yield', 'average_rain_fall_mm_per_year', 'pesticides_tonnes','avg_temp'

```python
# YOUR CODE HERE
subset_crops = crops[['hg/ha_yield', 'average_rain_fall_mm_per_year', 'pesticides_tonnes','avg_temp']]
#correlation matrix how much the fields are related to each other
```

```python
import seaborn as sns
```

```python
correlation_matrix = subset_crops.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix)
```

- Create a boxplot of the columns: 'hg/ha_yield', 'average_rain_fall_mm_per_year', 'pesticides_tonnes','avg_temp'

```python
# YOUR CODE HERE
plot = sns.boxplot(subset_crops)
plot.set_xticklabels(subset_crops.columns, rotation=45)
```

- Create a pairplot of the columns: 'hg/ha_yield', 'average_rain_fall_mm_per_year', 'pesticides_tonnes','avg_temp'

```python
# YOUR CODE HERE
sns.pairplot(subset_crops)
```

- The pairplot above has uncovered a potential problem with our data. Let's try to analize the distribution of the variables further to assess how grave the effect of skewness can be.

```python
# YOUR CODE HERE
#  log transform for the data
import numpy as np
subset_crops_log = subset_crops.apply(lambda x: np.log(x + 1))
subset_crops_log
```

```python
sns.pairplot(subset_crops_log)
```
