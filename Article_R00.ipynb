# **HETEROGENEITY**


Our initial analysis reveals no clear, systematic relationship between disaster-related deaths and international tourist arrivals. Similar disaster shocks are followed by very different tourism outcomes across countries, suggesting that disaster severity alone does not determine how tourists respond.

This becomes clearer if you imagine choosing your next holiday destination. You are rarely deciding between a country with disasters and a country without disasters. Instead, you compare destinations that may face similar risks but differ in how well disruptions are handled once they occur.

To capture this source of heterogeneity, we focus on the vulnerability component of the World Risk Index. While exposure measures how frequently and intensely countries are hit by natural hazards, vulnerability reflects how severely societies are affected when disasters occur. It summarizes structural features such as infrastructure quality, health and education systems, inequality, and institutional capacity. Figure X highlights the main dimensions underlying this vulnerability measure.

We conduct a vulnerability-based heterogeneity analysis by dividing countries into high-vulnerability and low-vulnerability groups using a baseline vulnerability score fixed at the start of the study period. We then estimate the relationship between disaster-related deaths and international tourist arrivals separately for each group.

This approach allows us to assess whether the same disaster shock leads to different tourism responses depending on a countrys underlying preparedness, isolating the role of structural vulnerability in shaping how destinations recover after natural disasters.
